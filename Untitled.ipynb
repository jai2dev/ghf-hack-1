```python
test_data["profession"]=pd.factorize(test_data.profession)[0]
test_data["city"]=pd.factorize(test_data.city)[0]
test_data["state"]=pd.factorize(test_data.state)[0]
test_data["married"]=pd.factorize(test_data.married)[0]
test_data["house_ownership"]=pd.factorize(test_data.house_ownership)[0]
test_data["car_ownership"]=pd.factorize(test_data.car_ownership)[0]


test_data=test_data.drop('state',axis=1)
test_data=test_data.drop('city',axis=1)
test_data=test_data.drop('married',axis=1)
test_data=test_data.drop('house_ownership',axis=1)
test_data=test_data.drop('car_ownership',axis=1)
test_data=test_data.drop('profession',axis=1)
result=pd.DataFrame()
result['id']=test_data['id']
```

```python
xtrain=train_data.drop("risk_flag",axis=1)
ytrain=train_data["risk_flag"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(xtrain,ytrain,test_size=0.2,random_state=101)
```

```python
num_folds = 10
num_instances = len(x_train)
seed = 7
scoring = 'roc_auc'

models = []
# models.append(('bagging',BalancedBaggingClassifier(base_estimator=DecisionTreeClassifier(class_weight= 'balanced',
#                                      max_depth= 70,
#                                      min_samples_leaf= 1,
#                                      min_samples_split= 4),
#                                  sampling_strategy=1,
#                                  replacement=False,
#                                  random_state=21)))
# models.append(('bagging',BalancedBaggingClassifier(base_estimator=RandomForestClassifier(bootstrap= ('True', 'False'),
#                          class_weight= 'balanced',
#                          max_depth= 82,
#                          min_samples_leaf= 7,
#                          min_samples_split= 7,
#                          n_estimators= 480),
#                                  sampling_strategy=1,
#                                  replacement=False,
#                                  random_state=21)))


models.append(('eec',EasyEnsembleClassifier(random_state=42)))

models.append(('rus',RUSBoostClassifier(random_state=0)))


models.append(('rr',BalancedRandomForestClassifier(max_depth=2, random_state=0, class_weight='balanced')))
# models.append(('bagging',BalancedBaggingClassifier(base_estimator=XGBClassifier(colsample_bytree= 0.8500000000000001,
#                          eta= 0.47500000000000003,
#                          eval_metric= 'auc',
#                          gamma= 0.8,
#                          max_depth= 7,
#                          min_child_weight= 5.0,
#                          n_estimators= 920,
#                          objective= 'binary:logistic',
#                          scale_pos_weight= 3.0,
#                          subsample= 0.8500000000000001),
#                                  sampling_strategy=1,
#                                  replacement=False,
#                                  random_state=21)))

# models.append(('XGB', XGBClassifier(colsample_bytree= 0.8500000000000001,
#                      eta= 0.17500000000000002,
#                      eval_metric= 'auc',
#                      gamma= 1.0,
#                      max_depth= 10,
#                      min_child_weight= 4.0,
#                      n_estimators= 470,
#                      objective= 'binary:logistic',
#                      scale_pos_weight=1.5,
#                      subsample= 0.7000000000000001)))
# models.append(('cat', CatBoostClassifier(bagging_temperature= 0.616823537464041,
#                  border_count= 224,
#                  depth= 5,
#                  iterations= 609,
#                  l2_leaf_reg= 2,
#                  learning_rate= 0.61,
#                  random_strength= 8.540000000000001,
#                  scale_pos_weight= 6.425819195353387)))
# models.append(('rr', RandomForestClassifier(bootstrap= ('True', 'False'),
#                          class_weight= 'balanced',
#                          max_depth= 82,
#                          min_samples_leaf= 7,
#                          min_samples_split= 7,
#                          n_estimators= 480)))

results = []
names = []
answers=[]
print("Scores for each algorithm:")
for name, model in models:
    kfold = KFold(n_splits = num_folds, random_state = seed, shuffle=True)
    cv_results = cross_val_score(model, x_train, y_train, cv = kfold, scoring = scoring)
    results.append(np.mean(cv_results))
    print(cv_results)
    names.append(name)
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    answers.append(roc_auc_score(y_test, predictions)*100)
    print(name, roc_auc_score(y_test, predictions)*100)
    print()
    print(classification_report(y_test,predictions))
```

```python
bagged_tree=BalancedBaggingClassifier(base_estimator=DecisionTreeClassifier(class_weight= 'balanced',
                                     max_depth= 70,
                                     min_samples_leaf= 1,
                                     min_samples_split= 4),
                                 sampling_strategy=1,
                                 replacement=False,
                                 random_state=12)

bagged_xgb=BalancedBaggingClassifier(base_estimator=RandomForestClassifier(bootstrap= ('True', 'False'),
                         class_weight= 'balanced',
                         max_depth= 82,
                         min_samples_leaf= 7,
                         min_samples_split= 7,
                         n_estimators= 480),
                                 sampling_strategy=1,
                                 replacement=False,
                                 random_state=12)

bagged_rrc=BalancedBaggingClassifier(base_estimator=XGBClassifier(colsample_bytree= 0.8500000000000001,
                         eta= 0.47500000000000003,
                         eval_metric= 'auc',
                         gamma= 0.8,
                         max_depth= 7,
                         min_child_weight= 5.0,
                         n_estimators= 920,
                         objective= 'binary:logistic',
                         scale_pos_weight= 3.0,
                         subsample= 0.8500000000000001),
                                 sampling_strategy=1,
                                 replacement=False,
                                 random_state=21)

cat=CatBoostClassifier(bagging_temperature= 0.616823537464041,
                 border_count= 224,
                 depth= 5,
                 iterations= 609,
                 l2_leaf_reg= 2,
                 learning_rate= 0.61,
                 random_strength= 8.540000000000001,
                 scale_pos_weight= 6.425819195353387)
```

```python
stack_gen = StackingCVClassifier(classifiers=( bagged_xgb,bagged_tree,bagged_rrc,cat),
                                meta_classifier=bagged_xgb,
                                use_probas=True,
                                use_features_in_secondary=True)
```

```python
stack_gen.fit(np.array(x_train),np.array(y_train))

p=stack_gen.predict(np.array(x_test))

print(roc_auc_score(y_test,p)*100)
print(classification_report(y_test,p))
```

```python
final = VotingClassifier(estimators=[('cat',cat),
        ('bt', bagged_tree), ('bx', bagged_xgb),('br', bagged_rrc), ('stack', stack_gen)],
        voting='soft', weights=[1.961,1.999,2.011,1.981,2.047],
        flatten_transform=True)
```

```python


final.fit(np.array(xtrain), np.array(ytrain))
```
